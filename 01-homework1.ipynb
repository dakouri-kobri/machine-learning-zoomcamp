Sum of weights

```python
# Asian cars
asian_cars = df[df['origin'] == 'Asia']
asian_cars.head()
```

```python
# Asian cars' vehicle_weight and model_year

asian_cars_select =  asian_cars[['vehicle_weight', 'model_year']]
asian_cars_select = asian_cars_select.head(7)
asian_cars_select
```

```python
# Turn asian_car_select into a NumPy array
X = asian_cars_select.to_numpy()
X
```

```python
# Matrix-matrix multiplication between the transpose of X and X
XTX = X.T @ X
XTX
```

```python
# Array y
y = [1100, 1300, 800, 900, 1000, 1100, 1200]
```

```python
# Compute w
m = np.linalg.inv(XTX) @ X.T
w = m @ y
w
```

```python
# Sum of all the elements of w
sum_w = w.sum()
sum_w
```
